```python
from collections import deque

dq1 = deque()
dq1.append(10)
dq1.append(20) 
dq1.append(30)
print(dq1)  
dq1.appendleft(5)
```

```python
d = deque()
print(d)
print(id(d))
print(type(d))
d.append(10)
print(d)
d.append(20)
print(d)
d.appendleft(65)
print(d)
d.appendleft(55)
print(d)
d.extend([20,30])
print(d)
d.extendleft([20,30])
print(d)
```

```python
food={'milk':100,'chips':50,'chocolates':200}
clothes={'shirts':150,'pants':100,'T-shirt':75}
electronic={'mobile':100,'laptop':50}

inv = {**food, **clothes, **electronic}
inv
inv['chips'] = 60
inv['shoes'] = 120
print(inv)
print(food)
```

```python
inv=ChainMap(food,clothes,electronic)
print(inv)
inv['milk']=50
print(inv)
print(food)
food['icecream']=100
print(inv)
inv['shoe']=25
print(inv)
clothes['shoe']=25
print(inv)
```
