### Single level inheritance

```python
class Parent:
    def m1(self):
        self.mobile = "Iphone"
        print("parent class method1")

class child(Parent):
    def m2(self):
        print("child class method2")

c = child()
c.m1()
c.m2()
print(c.mobile)
```

### multi level inheritance

```python
class GrandParent:
    def m1(self):
        print("Grand parent class method1")

class Parent(GrandParent):
    def m2(self):
        print("parent class method2")

class child(Parent):
    def m3(self):
        print("child class method3")

c = child()
c.m1()
c.m2()
c.m3()
```

            --> child1

            --> child2

### heirarchical inheritance

```python
class Parent:
    def m1(self):
        print("parent class method")

class child1(Parent):
    def cm1(self):
        print("child1 class method")

class child2(Parent):
    def cm2(self):
        print("child2 class method")

c1 = child1()
c1.cm1()
c1.cm2()
```

```python
class Father:
    def m1(self):
        print("Father class method")

class Mother:
    def m2(self):
        print("mother class method")

class child1(Father,Mother):
    def cm1(self):
        print("child class method")

c1 = child1()
c1.m1()
c1.m2()
c1.cm1()
```

```python
class GrandPa:
    def __init__(self):
        super().__init__()
        print("Grand Pa")

class Father(GrandPa):
    def __init__(self):
        super().__init__()
        print("Father")

class Mother(GrandPa):
    def __init__(self):
        super().__init__()
        print("Mother")

class child(Father, Mother):
    def __init__(self):
        super().__init__()
        print("Child")

c = child()
```
